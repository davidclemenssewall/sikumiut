```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
from datetime import timedelta
```

## Load data

```python
data_dir = '/glade/campaign/cgd/ppc/duvivier/sikumiut/DATA/fsnow/'
```

```python
hemi = 'NH'

# fsnow = 1.0
#fin = 'ice_state_timeseries_'+hemi+'_e.e21.E1850.f09_g17.CMIP6-piControl.001_branch2-0001-0050_tot'
#fout_base = 'ice_state_'+hemi+'_fsnow1.0_'

# all other fsnow (0.0, 0.25, 0.5, 0.75, 1.25, 1.5, 1.75)
value = str(0.0)
fin = 'ice_state_timeseries_'+hemi+'_seas_fsnow'+value+'_som-0001-0050_tot'
fout_base = 'ice_state_'+hemi+'_fsnow'+value+'_'
```

```python
ds_in = xr.open_mfdataset(data_dir + fin+'.nc')
```

```python
ds_in
```

## Average over all years - hs and aice

```python
ds_aice_avg = ds_in.aice.mean(dim='nyr')
ds_aice_std = ds_in.aice.std(dim='nyr')

ds_hs_avg = ds_in.hs.mean(dim='nyr')
ds_hs_std = ds_in.hs.std(dim='nyr')
```

## Save the timeseries

```python
fout = fout_base+'aice_avg'

# convert to dataframe
df = ds_aice_avg.to_dataframe()

# save file
df.to_csv(fout+'.csv')
```

```python
fout = fout_base+'aice_std'

# convert to dataframe
df = ds_aice_std.to_dataframe()

# save file
df.to_csv(fout+'.csv')
```

```python
fout = fout_base+'hs_avg'

# convert to dataframe
df = ds_hs_avg.to_dataframe()

# save file
df.to_csv(fout+'.csv')
```

```python
fout = fout_base+'hs_std'

# convert to dataframe
df = ds_hs_std.to_dataframe()

# save file
df.to_csv(fout+'.csv')
```
