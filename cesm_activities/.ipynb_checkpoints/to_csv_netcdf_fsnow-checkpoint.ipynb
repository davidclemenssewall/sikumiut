```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
from datetime import timedelta
```

## Load data

```python
data_dir = '/glade/campaign/cgd/ppc/duvivier/sikumiut/DATA/fsnow/'
```

```python
hemi = 'NH'
```

```python
# load each dataset

# all other fsnow (0.0, 0.25, 0.5, 0.75, 1.25, 1.5, 1.75)
value = str(0.0)
fin = 'ice_state_timeseries_'+hemi+'_seas_fsnow'+value+'_som-0001-0050_tot'
print(fin)
ds1 = xr.open_mfdataset(data_dir + fin+'.nc')

value = str(0.25)
fin = 'ice_state_timeseries_'+hemi+'_seas_fsnow'+value+'_som-0001-0050_tot'
print(fin)
ds2 = xr.open_mfdataset(data_dir + fin+'.nc')

value = str(0.5)
fin = 'ice_state_timeseries_'+hemi+'_seas_fsnow'+value+'_som-0001-0050_tot'
print(fin)
ds3 = xr.open_mfdataset(data_dir + fin+'.nc')

value = str(0.75)
fin = 'ice_state_timeseries_'+hemi+'_seas_fsnow'+value+'_som-0001-0050_tot'
print(fin)
ds4 = xr.open_mfdataset(data_dir + fin+'.nc')

fin = 'ice_state_timeseries_'+hemi+'_seas_e.e21.E1850.f09_g17.CMIP6-piControl.001_branch2-0001-0050_tot'
print(fin)
ds5 = xr.open_mfdataset(data_dir + fin+'.nc')

value = str(1.25)
fin = 'ice_state_timeseries_'+hemi+'_seas_fsnow'+value+'_som-0001-0050_tot'
print(fin)
ds6 = xr.open_mfdataset(data_dir + fin+'.nc')

value = str(1.5)
fin = 'ice_state_timeseries_'+hemi+'_seas_fsnow'+value+'_som-0001-0050_tot'
print(fin)
ds7 = xr.open_mfdataset(data_dir + fin+'.nc')

value = str(1.75)
fin = 'ice_state_timeseries_'+hemi+'_seas_fsnow'+value+'_som-0001-0050_tot'
print(fin)
ds8 = xr.open_mfdataset(data_dir + fin+'.nc')
```

```python
data_all = xr.concat([ds1,ds2,ds3,ds4,ds5,ds6,ds7,ds8],dim='expt')
```

```python
data_all
```

## Average over all years - hs and aice

```python
ds_aice_avg = data_all.aice.mean(dim='nyr')
ds_aice_std = data_all.aice.std(dim='nyr')

ds_hs_avg = data_all.hs.mean(dim='nyr')
ds_hs_std = data_all.hs.std(dim='nyr')
```

```python
# create figure
fig = plt.figure(figsize=(20,10))
# now loop through months to create subpanels
fout = 'fsnow_tests_'+hemi

mons = np.arange(1,13,1)

# Make subplots - note it's nrow x ncol x index (starting upper left)
# snow volume
ax = fig.add_subplot(1,2,1)
ax.plot(mons,ds_hs_avg.isel(expt=0),label="Fsnow_0.0",color='purple',linestyle='-',linewidth=2)
ax.plot(mons,ds_hs_avg.isel(expt=1),label="Fsnow_0.25",color='blue',linestyle='-',linewidth=2)
ax.plot(mons,ds_hs_avg.isel(expt=2),label="Fsnow_0.5",color='green',linestyle='-',linewidth=2)
ax.plot(mons,ds_hs_avg.isel(expt=3),label="Fsnow_0.75",color='gold',linestyle='-',linewidth=2)
ax.plot(mons,ds_hs_avg.isel(expt=4),label="Fsnow_1.0",color='black',linestyle='-',linewidth=2)
ax.plot(mons,ds_hs_avg.isel(expt=5),label="Fsnow_1.25",color='orange',linestyle='-',linewidth=2)
ax.plot(mons,ds_hs_avg.isel(expt=6),label="Fsnow_1.5",color='red',linestyle='-',linewidth=2)
ax.plot(mons,ds_hs_avg.isel(expt=7),label="Fsnow_1.75",color='pink',linestyle='-',linewidth=2)
# add text
plt.title('Snow volume',fontsize=15)
plt.ylabel('m^3',fontsize=15)
if hemi == 'NH':
    plt.legend(loc='upper right', fontsize=10, ncol=2)
else:
    plt.legend(loc='upper left', fontsize=10, ncol=2)   

# ice area
ax = fig.add_subplot(1,2,2)
ax.plot(mons,ds_aice_avg.isel(expt=0),label="Fsnow_0.0",color='purple',linestyle='-',linewidth=2)
ax.plot(mons,ds_aice_avg.isel(expt=1),label="Fsnow_0.25",color='blue',linestyle='-',linewidth=2)
ax.plot(mons,ds_aice_avg.isel(expt=2),label="Fsnow_0.5",color='green',linestyle='-',linewidth=2)
ax.plot(mons,ds_aice_avg.isel(expt=3),label="Fsnow_0.75",color='gold',linestyle='-',linewidth=2)
ax.plot(mons,ds_aice_avg.isel(expt=4),label="Fsnow_1.0",color='black',linestyle='-',linewidth=2)
ax.plot(mons,ds_aice_avg.isel(expt=5),label="Fsnow_1.25",color='orange',linestyle='-',linewidth=2)
ax.plot(mons,ds_aice_avg.isel(expt=6),label="Fsnow_1.5",color='red',linestyle='-',linewidth=2)
ax.plot(mons,ds_aice_avg.isel(expt=7),label="Fsnow_1.75",color='pink',linestyle='-',linewidth=2)
# add text
plt.title('Ice Area',fontsize=15)
plt.ylabel('m^2',fontsize=15)
if hemi == 'NH':
    plt.legend(loc='upper right', fontsize=10, ncol=2)
else:
    plt.legend(loc='upper left', fontsize=10, ncol=2)  

#add title
fig.suptitle(hemi+': Fsnow sanity test',fontsize=15)  
# save figure
fig = plt.savefig(fout+'.png', bbox_inches='tight', dpi=200)
```
